```python
# Python3 code to demonstrate working of
# Convert date range to N equal durations
# Using loop
import datetime

# initializing dates
startrange_ = datetime.datetime(2025, 9, 1)
endrange_ = datetime.datetime(2026, 1, 1)

             
# printing original dates
print("The start date is : " + str(startrange_))
print("The end date is : " + str(endrange_))

# initializing N
datediff = endrange_ - startrange_
N=(datediff/3).days

temp = []

# getting diff.
datediff = datediff // N
for idx in range(0, N):
    
    # computing new dates
    temp.append((startrange_ + idx * datediff))

# using strftime to convert to userfriendly 
# format
res = []
for sub in temp:
  res.append(sub.strftime("%Y-%m-%dT%H:%M:%S.%fZ"))


# printing result
print("N equal duration dates : " + str(res))
```
