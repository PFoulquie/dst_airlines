```python
startrange_ = "2026-01-01T00%3A00%3A00.000Z"
#endrange_ = "2026-01-31T23%3A59%3A59.000Z"
endrange_ = "2026-01-01T08%3A00%3A00.000Z"


# Premier appel pour avoir le nombre de pages 
response = requests.get(url=f"https://api.airfranceklm.com/opendata/flightstatus?startRange={startrange_}&endRange={endrange_}", headers=headers)
p_ = 0 
totalPages = json.loads(response.content.decode('utf-8'))["page"]['totalPages'] # Nombre total de pages 
print(f"Loop over {totalPages} pages")
list_af_get_flight = [f_ for f_ in  json.loads(response.content.decode('utf-8'))['operationalFlights']] 
p_ += 1 

# Puis loop sur les pages  
while p_ < totalPages:
    print(p_)
    response = requests.get(url=f"https://api.airfranceklm.com/opendata/flightstatus?startRange={startrange_}&endRange={endrange_}&pageNumber={p_}", headers=headers)

    try:
        list_af_get_flight.extend([f_ for f_ in  json.loads(response.content.decode('utf-8'))['operationalFlights']])
    except:
        print(f"Page {p_} failed with error {response.status_code}")
    p_ += 1 

```
